**CLEANING OUR DATASET**

**VALIDITY**



```python
#droping columns that are not useful for our study
# we drop the car column because it is redundant with Bluecar counter, always the same value
df = df.drop(['year', 'month', 'Cars'], axis = 1)
df.head()
```

**COMPLETENESS**

```python
# Checking if there is any missing value in dataframe as a whole
df.isnull()
```

```python
# dropping the columns with null values 
df = df.dropna(axis=1)
df.head()
```

```python
#confirming if null values have been dropped
df.isnull().sum()
```

```python
#checking the number of not null values in our data set
df.notnull().sum()
```

**IMPROVING CONSISTENCY**


```python
#checking for duplicates and dropping if any
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df.head()
```

**ENHANCING UNIFORMITY**

```python
#Maintaining uniformity in the column names
df.columns = df.columns.str.upper()
df.head()
```

DERIVING NEW ATTRIBUTES

```python
# Creating a new column where we find the difference in the number of bluecars at the station/counter

df['BLUECAR DIFFERENCE'] = df['BLUECAR COUNTER'].diff()
df['UTILIB DIFF'] = df['UTILIB COUNTER'].diff()
df['UTILIB 1.4 DIFF'] = df['UTILIB 1.4 COUNTER'].diff()
df = df.sort_values(["HOUR"], ascending=False)
df.head()
```

BLUE CAR DIFFERENCE IS NEGATIVE WHEN A CAR IS TAKEN

# DATA ANALYSIS


```python
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

city_filter = df[(df['CITY'] =='Paris') & (df['RENTAL STATUS'] == 'operational') & (df['BLUECAR DIFFERENCE'] < 0)]
city_filter

#Group by hour to find the most used bluecars
popular_hour = city_filter.groupby(['HOUR'])['BLUECAR DIFFERENCE'].sum().sort_values(ascending = True)
popular_hour.head()


#The most popular hour is 21. It has 193 cars
```

```python
#What is the most popular hour for returning cars?

return_cars = df[(df['RENTAL STATUS'] == 'operational') & (df['BLUECAR DIFFERENCE'] > 0)]
return_cars

return_cars_hour = return_cars.groupby(['HOUR'])['BLUECAR DIFFERENCE'].sum().sort_values(ascending = False)
return_cars_hour.head()

#hOUR 6 IS THE MOST POPULAR WITH 313 CARS RETURNED
```
